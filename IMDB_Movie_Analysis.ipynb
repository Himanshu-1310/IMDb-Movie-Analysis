## Data Cleaning

```python
# Convert date_x column from object to datetime

df['date_x'] = pd.to_datetime(df['date_x'])

df.info()
```

```python
# checking for duplicate

df[df.duplicated()]
```

```python
# Fill the null values in 'genre' and 'crew' column with 'N/A'

df['genre'] = df['genre'].fillna('N/A')
df['crew'] = df['crew'].fillna('N/A')

df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

## Univariate Analysis

### 1. What is the distribution of movie scores? Plot a histogram and describe its shape

```python
plt.figure(figsize=(10,6))
sns.histplot(df['score'],bins =20, kde=True, color = 'Yellow')
plt.title('Distribution of Movie Scores')
plt.xlabel('Score')
plt.savefig('Images/Distribution of Movie Scores.jpg')
plt.show()
```

The histogram is Left-Skewed(Negative Skewness) which means the data has more higher values and fewer lower values, with a few extreme small values.

#### 2. What is the most common genres in the IMDB? Use a bar chart to show their distribution.

```python
# Aggregate data of genre with total count of movies
gb = df.groupby('genre').agg({'names':'count'})
gb = gb.sort_values(by = 'names', ascending = False)
gb = gb.head(20)
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = gb.index, y = gb['names'], data = gb, palette = 'rocket')
plt.ylabel('Count of Movies')
plt.xlabel('Genre')
plt.title('Common Genres')
plt.xticks(rotation = 90)
plt.savefig('Images/Common Genre.jpg')
plt.show()
```

## Bivariate Analysis

#### 1.  Is there any relationship between a movie's budget and its revenue? Plot a scatter plot and describe any observed trend.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='budget_x', y = 'revenue', data=df, color ='black')
plt.title('Revenue vs Budget')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.savefig('Images/Revenue-Budget.jpg')
plt.show()
```

#### 2.How do scores vary by country? Use a boxplot to visualize the difference in scores across country.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='country',  y='score', data=df, color = 'orange')
plt.title('Score vs Country')
plt.xlabel('Country')
plt.ylabel('Score')
plt.xticks(rotation = 90)
plt.savefig('Images/Country-Score.jpg')
plt.show()
```

#### Is there a correlation between the score a movie received against its budget and revenue? Create a heatmap and calculate the correlation coefficient. 

```python
# Create a dataframe to store score, budget and revenue, and find its correlation.
df1 = df[['score','budget_x','revenue']]
df1.columns = ['Score','Budget','Revenue']

df1.head()
```

```python
corr = df1.corr()
corr
```

## Genre-Specific Analysis


#### 1. Which genre has the highest average score? Calculate the average score for each genre and plot the results.

```python
gb1 = df.groupby('genre').agg({'score':'mean'})
gb1 = gb1.sort_values(by='score',ascending = False)
gb1 = gb1.head()
```

```python
gb1
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=gb1.index, y= gb1['score'],data = gb1, palette='cubehelix')
plt.title('Average Score vs Genre')
plt.xlabel('Genre')
plt.ylabel('Average Score')
plt.xticks(rotation = 90)
plt.savefig('Images/Average Score-Genre.jpg')
plt.show()
```

#### How does the popularity of genres vary over time? Plot the number of movies released per genre each year.


```python
# Extract year from data_x column

df['year'] = df['date_x'].dt.strftime('%Y')
df['year'] = df['year'].astype(int)
df['year']
```

```python
#Group year and genre and count number of movie released each year in each group

gb2 = df.groupby(['year','genre']).size().reset_index(name='count')
gb2 = gb2.head(15)
```

```python
# Plot the data
plt.figure(figsize=(12,6))
sns.barplot(x='year',y='count',hue='genre', data = gb2, palette = 'Spectral')
plt.title('Count of Movies Per Genre Each Year')
plt.xlabel('Year')
plt.ylabel('Count of Movies')
plt.xticks(rotation = 90)
plt.legend(title='Genre',bbox_to_anchor = (1.05,1),loc = 'upper left')
plt.savefig('Images/Count of Movies Per Genre Each Year.jpg')
plt.show()
```

## Year and Trend Analysis

#### 1, How has the average score changed over the years? Plot the average score for each year.

```python
average_score  = df.groupby('year').agg({'score':'mean'})
average_score
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x=average_score.index, y = average_score['score'], data = average_score, palette = 'rainbow')
plt.title('Average Score over Years')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.xticks(rotation = 90, fontsize=6)
plt.savefig('Images/Average Score over Years.jpg')
plt.show()
```

#### 2. Which years had the highest and lowest number of movie releases? Plot the number of movies released each year.

```python
movies_count = df.groupby('year').agg({'names':'count'})
movies_count.tail(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = movies_count.index, y = movies_count['names'], data = movies_count, palette = 'plasma')
plt.title('Count of Movies Released')
plt.xlabel('Year')
plt.ylabel('Number of Movie Released each Year')
plt.xticks(rotation =90, fontsize = 6)
plt.savefig('Images/ Movies Released Each Year.jpg')
plt.show()
```

```python
# Which year has the highest and lowest movie release

year_count = df.year.value_counts()
year_count
```

```python
max_row = movies_count.loc[movies_count['names']== movies_count['names'].max()]
highest_release_year = max_row.index[0]


min_row = movies_count.loc[movies_count['names'] == movies_count['names'].min()]
lowest_release_year = min_row.index[0]

print(highest_release_year,',',lowest_release_year)
```

#### 3. Plot the number of movies released each decade

```python
df['decade'] = df['year'] // 10 * 10
```

```python
df['decade'].unique()
```

```python
decade = df.groupby('decade').agg({'genre':'count'})
decade
```

```python
plt.figure(figsize =(10,6))
sns.barplot(x= decade.index, y = decade['genre'],data = decade, palette = 'autumn')
plt.title('Movie Released Each Decade')
plt.xlabel('Year')
plt.ylabel('Count of Movie Released')
plt.xticks(rotation = 90)
plt.savefig('Images/Movies Released Each Decade.jpg')
plt.show()
```

# Insights and Summary


<b> Genre Popularity Over Time: </b> Certain genres, such as action and adventure, have seen a consistent rise in popularity over the years, like driven by advancements in special effects and global box office appeal. In contrast, genre like westerns or musicals have experienced a decline, possibly due to chnage in audience preferences and cultural trends.

<b> Impact of Budget on Movie Scores: </b> High budget movees often perform better in terms of audience and critic rating, as they can invest in better visual effects, renomved directors and top-tier actos. However, there are exceptions with some low-budget films (eg: independent dramas or thrillers) achieving critical acclaim due to strong storytelling and innovative filmmaking.

<b> Seasonal Release Trends and Scores: </b> Movies released during summer or holiday season tend to have higher box office earnings and audience rating. These periods are strategically chosen for blockbuster films that cater to family and mass audiences.
